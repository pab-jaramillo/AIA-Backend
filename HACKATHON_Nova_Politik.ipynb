```python
# ============================================
# CELDA 2: GENERADOR DE DATOS (NO MODIFICAR)
# ============================================
# Este c√≥digo crea TU pa√≠s √∫nico basado en tu ID

# Semilla basada en tu ID (garantiza datos √∫nicos)
np.random.seed(int(STUDENT_ID[-6:]))

# --- Crear geometr√≠a de 25 provincias (5x5) ---
provincias_nombres = [
    "Andina Norte", "Andina Centro", "Andina Sur", "Andina Este", "Andina Oeste",
    "Costa Norte", "Costa Centro", "Costa Sur", "Costa Este", "Costa Oeste",
    "Sierra Norte", "Sierra Centro", "Sierra Sur", "Sierra Este", "Sierra Oeste",
    "Amazonia Norte", "Amazonia Centro", "Amazonia Sur", "Amazonia Este", "Amazonia Oeste",
    "Insular Norte", "Insular Centro", "Insular Sur", "Insular Este", "Insular Oeste"
]

poligonos = []
for i in range(5):
    for j in range(5):
        poly = Polygon([(i, j), (i+1, j), (i+1, j+1), (i, j+1)])
        poligonos.append(poly)

# --- Crear datos con PROBLEMAS intencionales ---
data = {
    'provincia': provincias_nombres,
    'codigo': [f'NP-{str(i+1).zfill(2)}' for i in range(25)],
    'poblacion': np.random.randint(50000, 500000, 25),
    'protestas_reportadas': np.random.randint(0, 100, 25),
    'heridos': np.random.randint(0, 50, 25),
    'arrestos': np.random.randint(0, 200, 25),
    'desempleo_pct': np.random.uniform(5, 25, 25).round(1),
    'pobreza_pct': np.random.uniform(10, 60, 25).round(1),
}

# --- INSERTAR PROBLEMAS EN LOS DATOS ---

# Problema 1: Valores NULOS aleatorios
null_indices = np.random.choice(25, 5, replace=False)
for idx in null_indices[:2]:
    data['protestas_reportadas'] = list(data['protestas_reportadas'])
    data['protestas_reportadas'][idx] = np.nan
for idx in null_indices[2:4]:
    data['heridos'] = list(data['heridos'])
    data['heridos'][idx] = np.nan

# Problema 2: Una columna viene como STRING (deber√≠a ser n√∫mero)
data['arrestos'] = [str(x) if np.random.random() > 0.3 else x for x in data['arrestos']]

# Problema 3: Un OUTLIER extremo (provincia en crisis)
outlier_idx = np.random.randint(0, 25)
data['protestas_reportadas'] = list(data['protestas_reportadas'])
data['protestas_reportadas'][outlier_idx] = np.random.randint(800, 1200)
data['heridos'] = list(data['heridos'])
data['heridos'][outlier_idx] = np.random.randint(200, 400)

# Crear DataFrame
df_corrupto = pd.DataFrame(data)

# Crear GeoDataFrame
gdf_base = gpd.GeoDataFrame({'codigo': [f'NP-{str(i+1).zfill(2)}' for i in range(25)]},
                            geometry=poligonos)

print("="*60)
print("üåç REP√öBLICA DE NOVA-POLITIK - DATOS DE CRISIS")
print(f"   Versi√≥n generada para ID: {STUDENT_ID}")
print("="*60)
print("\nüìã Vista previa de los datos (CON PROBLEMAS):")
print(df_corrupto.head(10))
print("\n‚ö†Ô∏è ATENCI√ìN: Estos datos tienen ERRORES que debes corregir.")
```
